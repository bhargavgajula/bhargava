```python
new_data = df.drop(['Car_Name'], axis=1)
new_data['Current_year'] = 2023
new_data.head()
```

```python
new_data['number_of_year'] = new_data['Current_year'] - new_data['Year']
new_data.head()
```

```python
new_data.drop(['Year'],axis=1,inplace=True)
```

```python
new_data.head()
```

```python
new_data = pd.get_dummies(new_data,drop_first=True)
```

```python
new_data.head()
```

```python
new_data = new_data.drop(['Current_year'], axis=1)
new_data.head()
```

```python
import pandas as pd

# Your column names
column_names = ['Selling_Price', 'Present_Price', 'Kms_Driven', 'Owner',
                'number_of_year', 'Fuel_Type_Diesel', 'Fuel_Type_Petrol',
                'Seller_Type_Individual', 'Transmission_Manual']

# Creating an Index object
index_object = pd.Index(column_names)
```

```python
new_data.columns
```

```python
X = new_data.iloc[:,1:]
y = new_data.iloc[:, 0]
```

```python
from sklearn.ensemble import ExtraTreesRegressor

model = ExtraTreesRegressor()
model.fit(X, y)
```

```python
print(model.feature_importances_)
```

```python
feature_importances = pd.Series(model.feature_importances_, index=X.columns)
feature_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = RandomForestRegressor()
```

```python
random_grid = {'n_estimators': [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)],
               'max_features': ['auto', 'sqrt'],
               'max_depth': [int(x) for x in np.linspace(5, 30, num = 6)],
               'min_samples_split': [2, 5, 10, 15, 100],
               'min_samples_leaf': [1, 2, 5, 10]
            }
print(random_grid)
```

```python
model=RandomizedSearchCV(
    estimator = model,
    param_distributions = random_grid,
    scoring='neg_mean_squared_error',
    n_iter = 10, cv = 5, verbose=2,
    random_state=42, n_jobs = 1
)
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
data_frame = {
    'Actual_Value': y_test,
    'Prediction_Value':prediction
}

pred_vs_actu = pd.DataFrame(data_frame, columns=data_frame.keys())
pred_vs_actu.head()
```

```python
plt.figure(figsize=(5,3))
sns.distplot(y_test-prediction)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

def evaluate_metrics(y_test, prediction):
    print('MAE:', mean_absolute_error(y_test, prediction))
    print('MSE:', mean_squared_error(y_test, prediction))
    print('RMSE:', np.sqrt(mean_squared_error(y_test, prediction)))
```

```python
evaluate_metrics(y_test, prediction)
```
